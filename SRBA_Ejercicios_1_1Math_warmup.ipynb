```python
# Ejercicio #1
# Clase Math_warmup
# Sergio Raul Bonilla Alejo
# Fecha: 05 - Junio -2026

import numpy as np

def linear_1d_student(x, beta, omega):
    """
    Calcula y = beta + omega * x para x escalar o array-like.
    Devuelve un numpy.ndarray si x es array-like.
    """
    x = np.asarray(x)
    y = beta + omega * x
    return y

# Pruebas rápidas (no modificar)
x_test = np.array([0.0, 1.0, 2.0])
y_test = linear_1d_student(x_test, beta=2.0, omega=3.0)
assert np.allclose(y_test, np.array([2.0, 5.0, 8.0]))
print("All tests passed ✅")
```

```python
# MiniQuiz
#
# 1) Si ω = 0, ¿qué tipo de funcion se obtiene?
# Se obtiene una función constante, es decir una línea horizontal.
# Porque si ω = 0, la fórmula queda:
#
# 𝑦 = 𝛽 + 0 ⋅ 𝑥 = 𝛽
#
# La salida ya no depende de x.
# Esto coincide con la definición de intercepto y
# pendiente en la sección donde se explica que ω
# es la pendiente.
```

```python
# Grafica:
#
# La recta queda así:
# y = β
# ──────────────  ← línea horizontal
#
# No importa x, siempre vale β.
```

```python
# 2) ¿Cuál es el significado geométrico de β?
#
# β es el intercepto, es decir, el valor de y cuando x = 0.
# Geométricamente, es el punto donde la línea cruza el eje y.
```

```python
# Grafica:
#
# β = intercepto
# Es el punto donde la recta toca el eje y:
#
#   |
# β |●──────────
#   |
#   +----------------
```

```python
# 3) Si duplicas ω, ¿qué le pasa a la línea?
#
# Duplicar ω duplica la pendiente, así que la línea
# se vuelve el doble de inclinada:
# Si ω era positivo, la línea sube más rápido.
# Si ω era negativo, baja más rápido.
#
# El intercepto β no cambia, así que la línea
# sigue cruzando el eje y en el mismo punto.
#
# Esto se deduce de la explicación de que ω
# controla “qué tan rápido cambia y cuando
# cambia x”.
#
```

```python
# ω = 1   /         ω = 2   /
#        /                 /
#       /                 /
# La recta se vuelve más empinada.
```

```python
# MiniQuiz #4
#
# Si dos rectas tienen la misma ω pero diferente β, ¿se cruzan?
# No
# Son rectas paralelas, porque tienen la misma pendiente.
#
# Si dos rectas tienen diferente ω, ¿qué ocurre?
# Se cruzan en un punto, porque sus pendientes son distintas.
#
# ¿Qué determina el punto donde se cruzan?
#
# El punto de cruce depende de ambos valores β y ω.
# Se obtiene resolviendo:
#
# 𝛽1 + 𝜔1𝑥 = 𝛽2 + 𝜔2𝑥
#
```
